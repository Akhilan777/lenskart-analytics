```python
import requests
import pandas as pd
import re
```

```python
# To scrape contact lens data from Lenskart website
url = f'https://api-gateway.juno.lenskart.com/v2/products/category/3361?page-size=943&page=0'
res = requests.get(url)
data = res.json()
```

```python
product_list=data['result']['product_list']
df=pd.DataFrame(product_list)
df.to_csv('contact_lens_data.csv')
df.__len__()
```
